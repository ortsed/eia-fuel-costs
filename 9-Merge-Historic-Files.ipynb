```python
import pandas as pd
import os
```

```python
replace_map = {"ash": 'average_ash_content', "ash_content": 'average_ash_content',\
    "btu": "btu_content", "btucontent": "btu_content", "county_name": "county", \
    "generfuel": "gener_fuel", "plst_st": "plant_state", "plt_name": "plant_name", \
               "plt_st": "plant_state"
              }

def map_if(col):
    if col in replace_map:
        return replace_map[col]
    else:
        return col
    

def format_col_names(cols):



    cols = list(cols)
    cols = [col.lower() for col in cols]
    cols = [col.replace("\n", " ") for col in cols]
    cols = [col.replace(" ", "_") for col in cols]
    #cols = [col for col in cols if "blank" not in col]
    cols = [map_if(col) for col in cols]
    
    return cols
```
