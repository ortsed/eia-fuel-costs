```python
import pandas as pd
pd.set_option("display.max_columns", 500)
import os
```

```python
# function to format columns into a standard style, coalesce names
def format_column_name(cols):
    months = {"jan":"january", "feb":"february", "mar": "march", \
              "apr": "april", "may": "may", "jun": "june", \
              "jul": "july", "aug": "august", "sep":"september", \
              "oct": "october", "nov": "november", "dec":"december"}
    def replace_month(col):
        for  k,v in months.items():
            col = col.replace(v,k)
        return col
    
    cols = [col.replace("Elec_MMBtuSep", "elec_mmbtu_sep") for col in cols]
    cols = [col.lower() for col in cols]
    cols = [col.replace("\n", " ") for col in cols]
    cols = [col.replace(" ", "_") for col in cols]
    cols = [replace_month(col) for col in cols]
    
    cols = [col.replace("elec_mmbtus", "elec_mmbtu") for col in cols]
    cols = [col.replace("mmbtuper", "mmbtu_per") for col in cols]
    cols = [col.replace("mmbtu", "mmbtu_") for col in cols]
    cols = [col.replace(".1", "_1") for col in cols]
    cols = [col.replace(".2", "_2") for col in cols]
    cols = [col.replace("&", "and") for col in cols]
    cols = [col.replace("plant_state", "state") for col in cols]
    cols = [col.replace("reserved", "reserved_") for col in cols]
    cols = [col.replace("nuclear_unit_i.d.", "nuclear_unit_id") for col in cols]
    cols = [col.replace("elec_fuel_consumption_mmbtu_", "elec_fuel_consumption_mmbtu") for col in cols]
    cols = [col.replace("total_fuel_consumption_mmbtu_", "total_fuel_consumption_mmbtu") for col in cols]
    cols = [col.replace("elec_fuel_consumption_mmbtu_s", "elec_fuel_consumption_mmbtu") for col in cols]
    cols = [col.replace("total_fuel_consumption_mmbtu_s", "total_fuel_consumption_mmbtu") for col in cols]
    cols = [col.replace("elec_fuel_consumption_mmbtus", "elec_fuel_consumption_mmbtu") for col in cols]
    cols = [col.replace("total_fuel_consumption_mmbtus", "total_fuel_consumption_mmbtu") for col in cols]
    cols = [col.replace("net_generation_(megawatthours)", "net_generation_megawatthours") for col in cols]
    
    
    cols = [col.replace("__", "_") for col in cols]
    cols = [col.replace("__", "_") for col in cols]
    return cols
```
